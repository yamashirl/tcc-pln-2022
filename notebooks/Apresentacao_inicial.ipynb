```python
import time
import requests
from tika import parser
```

# Coleta da página

```python
# Dados constantes

url_base = 'http://diariooficial.imprensaoficial.com.br/doflash/prototipo/'
url_sufixo = '/exec1/pdf/pg_'

def pagina_pad(pagina):
    return str(pagina).zfill(4)

def monta_url(dia, mes, ano, pagina):
    return f'{url_base}{str(ano)}/{mes}/{str(dia)}{url_sufixo}{pagina_pad(pagina)}.pdf'

def monta_filename(dia, mes, ano, pagina):
    return f'{ano}{mes}{dia}_{pagina_pad(pagina)}'
```

**TODO:** Utilizar as ferramentas de
[`struct_time`](https://docs.python.org/3/library/time.html#time.struct_time) e
[`strftime()`](https://docs.python.org/3/library/time.html#time.strftime)
para utilizar dados numéricos nas variáveis.

```python
# Dados variáveis

dia = '14'
mes = 'abril'
ano = '2022'
pagina = 1
```

```python
# Download da página em um arq. pdf

while True:
    url_diario = monta_url(dia, mes, ano, pagina)
    filename = monta_filename(dia, mes, ano, pagina)
    pagina_padded = pagina_pad(pagina)
    req = requests.get(url_diario)

    if req.status_code == 200:
        with open(f'pdf/{filename}.pdf', 'wb') as pdf:
            pdf.write(req.content)
        print(f'Página {pagina_padded} baixada')
        
        raw = parser.from_file(f'pdf/{filename}.pdf')
        with open(f'txt/{filename}.txt', 'w') as texto_extraido:
            texto_extraido.write(str(raw['content']))
        print(f'Página {pagina_padded} extraída')
        
        time.sleep(2)
        pagina = pagina + 1
    elif req.status_code == 404:
        print('Final do diário/Página não encontrada')
        break
    elif req.status_code >= 500:
        print('Serviço Fora de Serviço')
        break
    else:
        print('Algo de errado não está certo')
        break
```
